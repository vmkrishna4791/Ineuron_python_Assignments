```python
#1. What does RGBA stand for?
#(A) A colour in an image is represented as an RGBA value. 
# An RGBA value is a set of numbers that indicates the amount of red, green, blue, and alpha (or transparency) in a color. 
```

```python
#2. From the Pillow module, how do you get the RGBA value of any images?
#!pip install Pillow
from PIL import Image
im=Image.new('RGBA',(50,50),'Blue')
im.save('blue_image.png') # Image blue_image.png is created in the local directory
flower_image = Image.open('blue_image.png') #The .jpg file should be current working directory
width,height=flower_image.size
print("Width=",width,"Height=",height)
rgba=flower_image.convert("RGBA")
values=rgba.getdata()
for item in values:
    print(item)  # Returns the RGBA values of the image blue_image.png

```

```python
#3. What is a box tuple, and how does it work?
#(A)
#Pillow functions take a box tuple argument.
#It is a tuple of four integer coordinates that represent a rectangular region in an image. 
# The four integers are, in order, as follows:

#Left - The x-coordinate of the leftmost edge of the Image box.

#Top - The y-coordinate of the top edge of the Image box.

#Right - The x-coordinate of one pixel to the right of the rightmost edge of the box. This integer must be greater than the left integer.

#Bottom - The y-coordinate of one pixel lower than the bottom edge of the box. This integer must be greater than the top integer.

```

```python
#4. Use your image and load in notebook then, How can you find out the width and height of an Image object?
#(A)
from PIL import Image
im=Image.new('RGBA',(50,50),'Yellow')
im.save('yellow_image.png') # Image yellow_image.png is created in the local directory
color_img = Image.open('yellow_image.png') #The .jpg file should be current working directory
width,height=color_img.size
print("Width=",width,"Height=",height)


```

```python
#5. What method would you call to get Image object for a 100×100 image, excluding the lower-left quarter of it?
#(A)
from PIL import Image, ImageDraw
new_im = Image.new('RGBA',(100,100),'Blue')
new_im.save('new_image.png') # new_image.png contains the 100x100 image which is in blue
new_im_cp=new_im.copy()
print(new_im_cp.size)
new_im_cp.getpixel((0, 0))
draw = ImageDraw.Draw(new_im_cp)
draw.rectangle((0, 49, 49, 99), fill='White')
new_im_cp.save('cropped_im.png') # cropped_im.png contains the lower left quarter excluded from it.

```
